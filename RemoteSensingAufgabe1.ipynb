```python
import pandas as pd
import sklearn
import matplotlib
import matplotlib.pyplot as plt
import numpy
import scipy
import os 
from pandas import read_csv
            
            
# Daten einlesen            
df_afrika = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/AfrikaTable.csv')
df = pd.DataFrame(df_afrika)
df.columns = ['value', 'count', 'grad']

df = df.drop(columns='grad')

print(df)
df_asien = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/asientable.csv')
df_australien = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/australiaTable.csv')
df_europe = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/europatable.csv')
df_northamerica = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/northamericaTable.csv')
df_ozeanien = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/ozeanienTable.csv')
df_southamerica = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/southamerica_table.csv')
df_antarktiv = pd.read_csv('/Users/nelesassor/Desktop/MachineLearning/Untitled Folder 2/tables/antarktiv_table.csv')

```

```python
valueAfrika = df_afrika['value']
countAfrika = df_afrika['count']

valueAsien = df_asien['value']
countAsien = df_asien['count']

valueAustralien = df_australien['value']
countAustralien = df_australien['count']

valueEurope = df_europe['value']
countEurope = df_europe['count']

valueNA = df_northamerica['value']
countNA = df_northamerica['count']

valueOzeanien = df_ozeanien['value']
countOzeanien = df_ozeanien['count']


valueSA = df_southamerica['value']
countSA = df_southamerica['count']

valueAntarktiv = df_antarktiv['value']
countAntarktiv = df_antarktiv['count']


fig, axs = plt.subplots(3, 3, sharey=True, tight_layout=True)
fig.suptitle("Elevationprofiles of continents", size=14)
fig.delaxes(axs[2,2])

axs[0, 0].hist(countAfrika)
axs[0, 0].set_title("Afrika")
axs[0, 1].hist(countAsien)
axs[0, 1].set_title("Asia")
axs[0, 2].hist(countAustralien)
axs[0, 2].set_title("Australia")
axs[1, 0].hist(countEurope)
axs[1, 0].set_title("Europe")
axs[1, 1].hist(countNA)
axs[1, 1].set_title("North America")
axs[1, 2].hist(countOzeanien)
axs[1, 2].set_title("Oceania")
axs[2, 0].hist(countSA)
axs[2, 0].set_title("South America")
axs[2, 1].hist(countAntarktiv)
axs[2, 1].set_title("Antarctica")
fig.savefig('hist_continents.pdf')  
```

```python
plt.plot(valueAfrika,countAfrika,  marker='.')
plt.plot(valueAsien,countAsien,  marker='.')
plt.plot(valueAustralien,countAustralien,  marker='.')
plt.plot(valueEurope,countEurope,  marker='.')
plt.plot(valueNA,countNA,  marker='.')
plt.plot(valueOzeanien,countOzeanien,  marker='.')
plt.plot(valueSA,countSA,  marker='.')
plt.plot(valueAntarktiv,countAntarktiv,  marker='.')
plt.xlabel('Runoff potential')
plt.ylabel('Counts')
plt.title('Runoff potential of continents')
plt.legend(['Afrika', 'Asia', 'Australia', 'Europe', 'North America', 'Oceania', 'South America', 'Antarctica' ])
plt.savefig('hist_continents_lines.pdf')  

```
